Checking Male/Female ratio

```python
stu['sex'].value_counts()
```

```python
gen = stu['sex'].value_counts()
plt.bar(gen.index , gen.values , color = [ 'lightcoral' , 'skyblue' ] , linewidth = 2 , label = 'student performance data')
plt.xlabel('Gender')
plt.ylabel('Counts')
plt.title('Distribution of Gender')
plt.grid(color = 'gray' , linestyle = ':' , linewidth = 1)
plt.xticks([0,1] , ['F' , 'M'])
plt.show()
```

```python
plt.pie(gen.values , labels = gen.index , colors = ['skyblue' , 'lightcoral'] , autopct = '%1.1f%%')
plt.title('Distribution of Gender')
```

Checking health condition of students

```python
health = stu['health'].value_counts()
plt.bar(health.index , health.values , color = ['#62e610' , '#e6d410' , '#c2e610' , '#a83232' ,'#e69410'] , linewidth = 2 , label = 'student performance data')
plt.xlabel('Health condition')
plt.ylabel('Counts')
plt.title('Health condition of Student')

plt.xticks([1,2,3,4,5] , ['Critical' , 'Poor' , 'Fair' , 'Good' , 'Very Good'])
plt.show()
```

```python
stu['avggrade'] = (stu['G1'] + stu['G2'] + stu['G3'])/3
stu['avggrade'].describe()
```

```python
plt.hist(stu['avggrade'] ,color = 'blue' , bins = 10, edgecolor = 'black' )
plt.xlabel('Student Average Score (out of 20)')
plt.ylabel('Number of Students')
plt.title('Avg marks of Student')
#colors = ['#FF0000', '#E02000', '#C04000', '#A06000', '#808000' , '#60A000', '#40C000', '#20E000', '#00CC00', '#00FF00' ]
```
